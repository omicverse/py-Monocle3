# py-Monocle3 — function-by-function R⇄Py parity

`pymonocle3` re-exports the full `monocle3-python` namespace, which is itself a 1:1 port of R `monocle3`. The mapping below is the canonical R-to-Python translation; everything is identical on both sides of the Python wrapper.

## 2.1 Preprocessing

| R monocle3 | Python (pymonocle3) | Notes |
|---|---|---|
| `new_cell_data_set(X, ...)` | `anndata.AnnData(X=X)` | AnnData replaces S4 cell_data_set |
| `estimate_size_factors(cds)` | `pymonocle3.estimate_size_factors(adata)` | identical |
| `preprocess_cds(cds, method, num_dim)` | `pymonocle3.preprocess_cds(adata, num_dim=...)` | truncated PCA |
| `align_cds(cds, alignment_group)` | `pymonocle3.align_cds(adata, alignment_group=...)` | mutual NN batch correction |

## 2.2 Dimensionality reduction + clustering

| R | Python | Notes |
|---|---|---|
| `reduce_dimension(cds, reduction_method)` | `pymonocle3.reduce_dimension(adata, ...)` | UMAP via umap-learn |
| `cluster_cells(cds, resolution)` | `pymonocle3.cluster_cells(adata, resolution=...)` | Leiden via leidenalg |
| `clusters(cds)` | `pymonocle3.clusters(adata)` | per-cell cluster ID |
| `partitions(cds)` | `pymonocle3.partitions(adata)` | per-cell partition (graph connected component) |

## 2.3 Trajectory inference + ordering

| R | Python | Notes |
|---|---|---|
| `learn_graph(cds)` | `pymonocle3.learn_graph(adata)` | SimplePPT principal graph |
| `order_cells(cds, root_pr_nodes=...)` | `pymonocle3.order_cells(adata, root_pr_nodes=...)` | pseudotime via graph dist |
| `pseudotime(cds)` | `pymonocle3.pseudotime(adata)` | per-cell pseudotime |
| `principal_graph(cds)` | `pymonocle3.principal_graph(adata)` | principal-graph object |

## 2.4 Differential expression

| R | Python | Notes |
|---|---|---|
| `graph_test(cds, neighbor_graph)` | `pymonocle3.graph_test(adata, ...)` | Moran's I per gene |
| `top_markers(cds, group_cells_by)` | `pymonocle3.top_markers(adata, group_cells_by=...)` | per-cluster marker genes |
| `find_gene_modules(cds)` | `pymonocle3.find_gene_modules(adata)` | UMAP + Leiden on gene-gene graph |
| `aggregate_gene_expression(cds, modules)` | `pymonocle3.aggregate_gene_expression(adata, modules)` | module-level expression |

## 2.5 Visualisation

| R | Python | Notes |
|---|---|---|
| `plot_cells(cds, color_cells_by=...)` | `pymonocle3.plot_cells(adata, color_cells_by=...)` | returns ggplot2-python plot |
| `plot_cells_3d(cds, ...)` | `pymonocle3.plot_cells_3d(adata, ...)` | plotly 3D scatter |
